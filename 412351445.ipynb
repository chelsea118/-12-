```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd


url = "https://www.taifex.com.tw/cht/3/futDailyMarketReport?"
payload ={
    "queryType":2,
    "marketCode":0,
    "commodity_id":"TX",
    "queryDate":"2025/10/08",
    "MarketCode":0,
    "commodity_idt":'MTX'

}

payload["queryDate"] = "2025/10/08"
res = requests.get(url, params = payload)
soup = bs(res.text, "lxml")
tb = soup.select("table")[0]
df = pd.read_html(str(tb))[0]
df
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd


url = "https://www.taifex.com.tw/cht/3/futDailyMarketReport?"
payload ={
    "queryType":2,
    "marketCode":0,
    "commodity_id":"TX",
    "queryDate":"2025/10/08",
    "MarketCode":0,
    "commodity_idt":'MTX'

}

payload["queryDate"] = "2025/10/08"
res = requests.post(url, data = payload)
soup = bs(res.text, "lxml")
tb = soup.select("table")[0]
df = pd.read_html(str(tb))[0]
df
```
